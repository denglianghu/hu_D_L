```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import matplotlib as mpl
%matplotlib inline
```

```python
#正常显示中文
mpl.rcParams['font.sans-serif']=[u'SimHei']
mpl.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv("D:\datasets\Pokemon.csv",index_col = 0,sep = ",")
```

```python
sns.lmplot(x = "Attack",y = "Defense",data = df,fit_reg = False,hue = "Stage")
plt.xlim(0,None)
plt.ylim(0,None)
plt.show()
```

```python
sns.boxplot(data = df)
```

```python
stage_df = df.drop(["Total","Stage","Legendary"],axis = 1)
sns.boxplot(data = stage_df)
```

```python
df.describe()
```

```python
sns.set_style("whitegrid")
sns.violinplot(x="Type 1",y="Attack",data = df)
plt.xlabel("Type 1")
plt.ylabel("Attack")
plt.show()
```

```python
sns.swarmplot(x="Type 1",y="Attack",data = df)
```

```python
plt.figure(figsize = (10,6))
sns.violinplot(x="Type 1",y="Attack",data = df,inner = None)
sns.swarmplot(x="Type 1",y="Attack",data = df,color = "k",alpha = 0.7)
plt.title("Attack by Type")
plt.show()
```

```python
pd.options.display.max_rows = 10
stage_df
```

```python
melted_df = pd.melt(stage_df,id_vars = ["Name","Type 1","Type 2"],var_name = "Stat")
melted_df
```

```python
plt.figure(figsize = (10,6))
sns.swarmplot(x = "Stat",y="value",data = melted_df,hue = "Type 1")
```

```python
plt.figure(figsize = (10,6))
sns.swarmplot(x = "Stat",y="value",data = melted_df,hue = "Type 1",split=True)
plt.ylim(0,260)
plt.legend(bbox_to_anchor = (1,1),loc = 2)
```

```python
#Heatmap可以帮助可视化矩阵状的数据。
sns.heatmap(stage_df.corr())
```

```python
#Histogram能够绘制变量的数值分布。 柱状图
sns.distplot(df["Attack"])
```

```python
#绘制条形图  barplot
sns.countplot(x = "Type 1",data = df)
```

```python
#分离图表
g = sns.factorplot(x = "Type 1",
            y = "Attack",
            data = df,
            hue = "Stage",#根据颜色来分离图标
            col = "Stage",#根据Stage来分离图标
            kind = "swarm")
g.set_xticklabels(rotation = -45)
```

```python
#绘制密度图
sns.kdeplot(df.Attack,df.Defense)
#曲线越密集的地方说明两个变量的关系越近，越稀疏的地方说明关系越远。 
```

```python
#联合分布图将散点图和直方图的信息结合起来，提供双变量分布的详细信息。 
sns.jointplot(x = "Attack",y = "Defense",data = df)
```
